#converting the values from '0' and '1' to 'Yes' and 'No' in 'SeniorCitizen'column to make it easier to understand

```python
def conv(value):
    if value==1:
        return "Yes"
    else:
        return "No"

df["SeniorCitizen"]=df["SeniorCitizen"].apply(conv)
```

```python
df.head(30)
```

```python
ax=sns.countplot(x='Churn',data=df)
ax.bar_label(ax.containers[0]) #The x-axis will show "Yes" and "No"
plt.title("Count of Customers By Churn")
plt.show()
```

```python
#Creating a pie chart for customers who have Churned out
```

```python
plt.figure(figsize=(3,4))
gb=df.groupby('Churn').agg({'Churn':"count"})
plt.pie(gb['Churn'],labels=gb.index,autopct="%1.2f%%")
#labels=gb.index sets the labels as 'Yes' and 'No'.
#autopct="%1.2f%%" shows percentages on the pie slices with 2 decimal places 
plt.title("Percentage of Churn customers")
plt.show()
```

#from the above pie chart we can conclude that 26.45% of our customers have churned out.


```python
plt.figure(figsize=(3,3))
ax=sns.countplot(x="gender",data=df,hue="Churn")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title("Churn by Gender")
plt.show()
```
